# imports

```python
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

# Create Data Frame

```python
housing_data={
    'No_of_rooms':[2,3,4,3,2,None,4,3,None,2],
    'No_of_Bathrooms':[1,2,3,2,1,2,None,2,3,None],
    'Area_in_Sqfit':[800,1200,1500,None,900,1000,2000,1800,None,750],
    'Location':['Urban','Suburban','Urban','Rural','Urban','Suburban','Rural','Urban','Suburban',None],
    'Price':[500000,650000,850000,400000,520000,600000,1000000,950000,700000,None]
}
```

```python
housing_df=pd.DataFrame(housing_data)
```

```python
housing_df
```

```python
housing_df.shape
```

```python
housing_df.info()
```

```python
housing_df.describe()
```

# Handle Missing Values

```python
housing_df.isna().sum()
```

```python
housing_df['No_of_rooms']=housing_df['No_of_rooms'].fillna(math.ceil(housing_df['No_of_rooms'].mean()))
housing_df
```

```python
housing_df['No_of_Bathrooms']=housing_df['No_of_Bathrooms'].fillna(housing_df['No_of_Bathrooms'].median())
housing_df
```

```python
housing_df['Area_in_Sqfit']=housing_df['Area_in_Sqfit'].fillna(housing_df['Area_in_Sqfit'].mode()[0])
housing_df
```

```python
housing_df['Price']=housing_df['Price'].fillna(housing_df['Price'].max())
housing_df
```

# Encoding

```python
object_cols=housing_df.select_dtypes(include=object).columns
object_cols
```

```python
housing_df=pd.get_dummies(housing_df,columns=object_cols)
```

```python
housing_df
```

```python
bool_cols=housing_df.select_dtypes(include=bool).columns
bool_cols
```

```python
housing_df[bool_cols]=housing_df[bool_cols].astype(int)
```

```python
housing_df
```

# Splitting

```python
Y=housing_df['Price']
X=housing_df.drop(columns=['Price'])
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```
