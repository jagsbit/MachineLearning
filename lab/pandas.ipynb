```python
housing_df=housingFrame.copy()
housing_df
```

```python
import math
```

```python
housing_df['No_of_rooms']=housing_df['No_of_rooms'].fillna(math.floor(housing_df['No_of_rooms'].mean()))
```

```python
housing_df
```

```python
housing_df.isna().sum()
```

```python
housing_df['No_of_Bathrooms']=housing_df['No_of_Bathrooms'].fillna(housing_df['No_of_Bathrooms'].mode()[0])
```

```python
housing_df
```

```python
housing_df['Area_in_sqfit']=housing_df['Area_in_sqfit'].fillna(housing_df['Area_in_sqfit'].median())
```

```python
housing_df
```

```python
housing_df=pd.get_dummies(housing_df,columns=['Location'])
```

```python
housing_df
```

```python
bool_cols=housing_df.select_dtypes(include=bool).columns
```

```python
bool_cols
```

```python
housing_df[bool_cols]=housing_df[bool_cols].astype(int)
```

```python
housing_df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
columns_to_normalize = ['Area_in_sqfit']
```

```python
normalizer=MinMaxScaler()
housing_df[columns_to_normalize]=pd.DataFrame(normalizer.fit_transform(housing_df[columns_to_normalize]),columns=columns_to_normalize)
housing_df
```
